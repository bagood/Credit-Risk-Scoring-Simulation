# Import Packages

```python
from warnings import simplefilter
simplefilter('ignore')

import cloudpickle
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix, log_loss

pd.set_option('display.max_columns', 999)
```

# Load Data and Model

```python
train_path = Path("train")
X_train = pd.read_csv((train_path / 'X').with_suffix('.csv'))
y_train = pd.read_csv((train_path / 'y').with_suffix('.csv'))
```

```python
test_path = Path("test")
X_test = pd.read_csv((test_path / 'X').with_suffix('.csv'))
```

```python
model_path = Path('model')

with open((model_path / 'lgb_model').with_suffix('.pkl'), 'rb') as f:
    model = cloudpickle.load(f)
```

```python
model
```

# Model Performance

Because we are focused on developing a credit risk scoring model, the main metric that we will be observing is the ROC AUC

```python
class_1_index = np.where(model.classes_ == 1)[0][0]

train_proba_predictions = model.predict_proba(X_train[model.feature_name_])[:, class_1_index]
train_predictions = model.predict(X_train[model.feature_name_])
```

Having an AUC of 0.728 means that the model is quite capabale in separating the model's prediction probability between the "good" and "bad" individuals

```python
train_auc = roc_auc_score(y_train, train_proba_predictions)
print(f"Training AUC: {train_auc}")
```

```python
train_confusion_matrix = confusion_matrix(y_train, train_predictions)
train_confusion_matrix_df = pd.DataFrame(
    train_confusion_matrix, 
    index=[f'True: {i}' for i in [0, 1]], 
    columns=[f'Pred: {i}' for i in [0, 1]]
)
print("Training Confusion Matrix \n")
train_confusion_matrix_df
```

```python
train_log_loss = log_loss(y_train, train_proba_predictions)
print(f"Training Log Loss: {train_log_loss}")
```

```python
print('Train Classification Report')
print(classification_report(y_train, train_predictions))
```

# Model Stability

Since we have no labels over the testing data, we can still measure the capability of our model using the PSI value. PSI value is also very important for our model as it provides information of whether we need to retrain our model or cases where there is a great shift in distribution of the features used by the model

```python
def psi(ref, new, bins=10):
     
    ref, new = np.array(ref), np.array(new)
    
    quantiles = np.linspace(0, 1, bins + 1)
    breakpoints = np.quantile(ref, quantiles)
    
    ref_counts = np.histogram(ref, breakpoints)[0]
    new_counts = np.histogram(new, breakpoints)[0]
    
    ref_pct = ref_counts / len(ref)
    new_pct = new_counts / len(new)
    
    ref_pct = np.where(ref_pct == 0, 1e-6, ref_pct)
    new_pct = np.where(new_pct == 0, 1e-6, new_pct) 
    
    return np.sum((ref_pct - new_pct) * np.log(ref_pct / new_pct))
```

```python
class_1_index = np.where(model.classes_ == 1)[0][0]

test_proba_predictions = model.predict_proba(X_test[model.feature_name_])[:, class_1_index]
test_predictions = model.predict(X_test[model.feature_name_])
```

```python
psi_score = psi(train_proba_predictions, test_proba_predictions)

all_psi_feature = {}

for feature in model.feature_name_:
    psi_feature = psi(X_train[feature].values, X_test[feature].values)
    all_psi_feature[feature] = [psi_feature]
```

```python
psi_df = pd.concat(
                    (
                        pd.DataFrame(all_psi_feature).T \
                            .reset_index() \
                            .rename(columns={'index': 'feature', 0: 'psi'}),
                        pd.DataFrame({
                            'feature': ['default_probability_score'],
                            'psi': [psi_score]
                        })
                    )
                )
```

Here we can see that the PSI value is way below 0.1, meaning that the distribtion of the features between the training and testing data is quite similar

Observe that the PSI value for the model's prediction (default_probability_score) is very low, meaning that the performance of our model tested on the training data would not be greatly differs if we were to test it on the testing data

```python
psi_df.sort_values('psi', ascending=False)
```

# Model Explainability

In the banking industry, having a model that is explainable is at upmost important as we need to have strong justification over stating that an individual is considered "good" or "bad"

We can easily explain this by lookin at the SHAP ratio of our model

```python
import shap

explainer = shap.Explainer(model, X_train)

shap_values = explainer(X_test)
```

```python
shap.summary_plot(shap_values, X_test)
```
