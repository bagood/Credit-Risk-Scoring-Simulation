Hi, this is the first of three sections of developing a credit risk scoring model. In this section, we will dive in to the kinds of data that we have, what informations it contains, and how we can leverage the data to create a good credit risk scoring model

## What did you discover from the datasets given?

I discover that the data contains information related to historical records of a set of population that has taken a loan. Individuals in this population that has taken a loan are labeled as "good" and "bad", likely based on a certain DPD (Days Past Due) of their loan payment. This data will be used for developing a credit risk scoring model, aiming to lower the probability of giving a loan to an individual that have strong probability of not paying their loan in time.

One very insightful findings that I would like to highlight is that the value "good_bad" of 1 is the majority class, while 0 is the minority class. In the world of credit risk scoring, the label should be otherwise. This insight leads me to flip the label between 0 and 1 to have a proper data for model development

The data contains several continuous and categorical variables that could be useful in the development of the credit risk scoring model

An assumption is made for the column "acc_now_delinq" that contain missing values. Based on the definition of the column, missing values could mean that the individual have not yet taken a loan on the individual's account. Hence, I impute it with the value -1

After looking at the distribution of the data, it would appear that there are some outliers. But, these outlier are unlikely error, rather a naturally occuring events. Especially for the value "annual_inc" which is very an individual to have an annual income of that amount

![Alt Text](figure/distribution_of_continuous_variables.png)

Using one continuous variable alone to predict whether an indivudal is considered "good" or "bad" would not be a good approach as the distribution of the continuous variables over the label is quite overlapping

![Alt Text](figure/distribution_of_continuous_variables_over_lables.png)

The same thing goes with the categorical variables

![Alt Text](figure/percentage_of_good_bad_on_categorical_variables.png)

One unique insights that i found is that none of the continuous variables correlated with one another. This could be a great start for model development as the variables could provide clear pattern of our data during model developments

![Alt Text](figure/correlation_continuous_variables_heatmap.png)

## How strong is your predictive power of your independent variable?

The predictive power of the independent variable is relatively small. We can see this through the distribution of the continuous and categeorical variables have over the label. # Get IV and WOE

To have a better overview of the predictive power for each variables, we will be using IV (Information Value)

Variables "int_rate" and "grade" both shows a very strong predictive power, shown by the value of their IV exceeding 0.3

But, the rest of the variables shows a weak predictive power. This could be an indication for the need of new variables that could provide better predictive power

```python
dict_cat_binning_woe_results = {}

for var in categorical_variables:
    woe_results = categorical_woe(train_data.copy(), 
                                  cat_variable_name=var,
                                  y_df=train_data[target_column])

    dict_cat_binning_woe_results[var] = woe_results

dict_cat_iv = {key: val['total_IV'].unique() for key, val in dict_cat_binning_woe_results.items()}
```

```python
list_cont_binning_woe_results = {}

for var in continuous_variables:
    _, _, binning_woe_results = continuous_woe(train_data.copy(), 
                                                    feature = var, 
                                                    target = target_column, 
                                                    max_bins=20, 
                                                    min_samples_bin=0.05)

    list_cont_binning_woe_results[var] = binning_woe_results

dict_cont_iv = {key: val['total_optimized_IV'].unique() for key, val in list_cont_binning_woe_results.items()}
```

```python
all_iv = pd.concat(
    (
        pd.DataFrame(dict_cat_iv).T,
        pd.DataFrame(dict_cont_iv).T,
    )
).rename(columns={0: 'IV'}) \
    .sort_values('IV', ascending=False)
```

```python
all_iv
```

# Prepare Training Data

We will prepare our training and testing data based on our previous findings. First, we will flip the labels, 0 will be flipped into 1 and 1 will be flipped into 0. Second, we will impute the mssing values on variable "acc_now_delinq" into -1 as it could represent that the person have no activities related with the variable. Lastly, we will perform one-hot encoding on the categorical variables

```python
def prepare_data(data, continuous_variables, categorical_variables, target_column):
    if target_column != None:
        data[target_column] -= 1
        data[target_column] *= -1

    data['acc_now_delinq'].fillna(-1, inplace=True)

    prep_data = pd.get_dummies(data, columns=categorical_variables)
    prep_data.columns = [col.replace(' ', '_') for col in prep_data]
    
    bool_var = [var for var, var_type in prep_data.dtypes.to_dict().items() if var_type == bool]
    for var in bool_var:
        prep_data[var] = prep_data[var].astype(int)

    return prep_data
```

```python
train_data = pd.read_csv('Data Science Test/pd_loan_data_train.csv')
test_data = pd.read_csv('Data Science Test/pd_loan_data_test.csv')

train_data.drop(columns=['loan_id'], inplace=True)
test_data.drop(columns=['loan_id'], inplace=True)
```

```python
prep_train_data = prepare_data(train_data, continuous_variables, categorical_variables, target_column)
prep_test_data = prepare_data(test_data, continuous_variables, categorical_variables, None)
```

```python
assert len(prep_train_data.columns) - 1 == len(prep_test_data.columns)

feature_columns = prep_train_data.columns.tolist()
feature_columns.remove('good_bad')
```

```python
def write_prepared_data(data, feature_columns, target_column, data_path):
    data_path = Path(data_path)
    data_path.mkdir(exist_ok=True)
    
    X = data[feature_columns]
    X.to_csv((data_path / 'X').with_suffix('.csv'), index=False)

    if target_column != None:
        y = data[[target_column]]
        y.to_csv((data_path / 'y').with_suffix('.csv'), index=False)

    return
```

```python
write_prepared_data(prep_train_data, feature_columns, target_column, 'train')
write_prepared_data(prep_test_data, feature_columns, None, 'test')
```
