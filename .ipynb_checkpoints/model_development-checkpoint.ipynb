# README

Welcome to the second section of a credit risk scoring model. This section will be about the process of developing the model

In this section, a total of 4 different models are used as candidates for selecting the best model that best suited the pattern of our data

After we select the best model, this model went through 2 model optimization process, which are hyperparameter tuning and feature selection process

Finally, a model is developed based on the result of the experiments and optimizations

## How did you optimized the model?

To optimize the model, I went on performing experiment on 4 different models (Logistic Regression, XGBoost, LightGBM, and CatBoost) to have the best fitted model based on the patterns of our data. These models are trained using the default parameters to get a sense of their raw capabilities. The final selected model that outperforms the other model is LightGBM. The model have best balance between maintining good fit over the data and generalization over unseen data

LightGBM then went through a hyperparameter tuning process to get the best parameters based on the patterns of our data. The hyperparameter tuning is done using the grid search cross validation method

After the best parameters has been acquired, I found that there are several features that has weak to no importance over the model. Hence, I perform a recursive feature elimination to ensure that features used by the model is not a noise

Finally, we train LightGBM over the acquired best parameters using the selected features over the entire training data. This ensures that the model have all the knowledge to generalize over the testing data

## How did you make sure the model is stable enough?

To ensure that the model is stable we need to make sure the model used is best suited for our data, the parameters are tuned to maximized performance, and eliminate features that are considered as noise by the model (have weak to no feature importance)

Using less features also means that the total amount of features that we need to monitor using the PSI value is
also lessen

## Did you use ensemble method to improve model performance?

Yes, LightGBM is a machine learning algorithms that uses the ensemble method. This model proves to have bettern performance over linear model like Logistic Regression

# Import Packages

```python
from warnings import simplefilter
simplefilter('ignore')

import pandas as pd
import numpy as np
import xgboost as xgb
import catboost as cb
import lightgbm as lgb
from pathlib import Path
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, classification_report
from sklearn.model_selection import GridSearchCV, StratifiedKFold

pd.set_option('display.max_columns', 999)
```

# Load Training Data

```python
train_path = Path("train")
X = pd.read_csv((train_path / 'X').with_suffix('.csv'))
y = pd.read_csv((train_path / 'y').with_suffix('.csv'))
```

```python
X
```

```python
y['good_bad'].value_counts('good_bad')
```

# Select Best Performing Model

```python
def evaluate_model_cv(model_name, model, X, y, n_splits=5):
    """
    Evaluate model performance on each of the strattified k-fold cross validation
    """
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    
    train_aucs = []
    val_aucs = []
    
    X_arr = np.array(X)
    y_arr = np.array(y)
    
    for train_index, val_index in skf.split(X_arr, y_arr):
        X_train_fold, X_val_fold = X_arr[train_index], X_arr[val_index]
        y_train_fold, y_val_fold = y_arr[train_index], y_arr[val_index]
        
        model.fit(X_train_fold, y_train_fold)
        
        class_1_index = np.where(model.classes_ == 1)[0][0]
        train_preds = model.predict_proba(X_train_fold)[:, class_1_index]
        val_preds = model.predict_proba(X_val_fold)[:, class_1_index]
        
        train_aucs.append(roc_auc_score(y_train_fold, train_preds))
        val_aucs.append(roc_auc_score(y_val_fold, val_preds))

    return pd.DataFrame({
        'Model': [model_name],
        'Train ROC AUC (Mean)': [np.mean(train_aucs)],
        'Val ROC AUC (Mean)': [np.mean(val_aucs)]
    })

def compare_all_models(X, y):
    """
    Orchestrate the process of batch training and measuring of model
    """
    models = {
        'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
        'XGBoost': xgb.XGBClassifier(eval_metric='auc', random_state=42),
        'LightGBM': lgb.LGBMClassifier(metric='auc', verbose=-1, random_state=42),
        'CatBoost': cb.CatBoostClassifier(eval_metric='AUC', verbose=0, random_state=42)
    }

    results = []

    for name, model in models.items():
        model_results = evaluate_model_cv(name, model, X, y, n_splits=5)
        results.append(model_results)

    final_results_df = pd.concat(results, ignore_index=True)
    
    return final_results_df
```

```python
all_roc_auc = compare_all_models(X, y)
```

The goal of this step is to select a model that could best fit the patterns of our data. This is a very crucial step because putting lots of effort on a model that doesn't fit our data would result in a bad model performance

We will evaluate each of the model using the ROC AUC value as this is a common metric used for measuring a model's performance. ROC AUC explains how well the model can distinguish between "good" and "bad" individuals through the model's resulted prediction probability. Prediction probability is our main focus because we would be using that value for generating the score card for our credit scoring model.

Based on the result, LightGBM shows the best capability in learning the pattern of our training data while remaining good capabilites on data that it has not seen yet (validation data)

Although CatBoost shows the best performnance in the validation data compared to LightGBM, it loses quite a lot in the performance of the training data

```python
all_roc_auc
```

# Hyperparameter Tuning on Best Performing Model (LightGBM)

Now that we have selected LightGBM as the best performing model, we will find the best suited parameters for the patterns of our data

```python
lgb_model = lgb.LGBMClassifier()

param_grid = {
    'random_state': [42], 
    'metric': ['auc'], 
    'is_unbalance': [True], 
    'n_jobs': [-1],
    'max_depth': [3, 5], 
    'num_leaves': [7, 31], 
    'min_child_samples': [100, 250, 500], 
    'reg_alpha': [0.1, 10] 
}

cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

lgb_grid_search = GridSearchCV(
    estimator=lgb_model,
    param_grid=param_grid,
    scoring='roc_auc',
    cv=cv_strategy,
    verbose=2,
    n_jobs=-1
)

lgb_grid_search.fit(X, y)
```

We have already seeing an increase to the performance of our model

```python
best_lgb = lgb_grid_search.best_estimator_
best_lgb_params = lgb_grid_search.best_params_

print(f"Best AUC: {lgb_grid_search.best_score_}")

class_1_index = np.where(best_lgb.classes_ == 1)[0][0]
proba_predictions = best_lgb.predict_proba(X)[:, class_1_index]
predictions = best_lgb.predict(X)

final_auc = roc_auc_score(y, proba_predictions)
print(f"Final AUC: {final_auc}")

print('Classification Report')
print(classification_report(y, predictions))
```

Now, we will be looking at the feature importance to see whether there are features that shows weak to no importance to our model. We would need to remove these features as it could introduce unwanted noise to our model

We can see that there are quite a lot of features with 0 importance

```python
feature_importance_df = pd.DataFrame(
    {
        'Feature': best_lgb.feature_name_,
        'Importance': best_lgb.feature_importances_
    }
)
```

```python
feature_importance_df.sort_values('Importance', ascending=False).reset_index(drop=True)
```

# Recursive Feature Elimination CV on LightGBM with it's Best Performing Parameters

In this step, we will be removing less important features from our model using the Recursive Feature Elimination CV. This would remove these less important features incorporating cross validation

As for the parameters, we will be using parameters that resulted in the best model performance based on the previous step

```python
estimator = lgb.LGBMClassifier(**best_lgb_params)

cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

selector = RFECV(
    estimator=estimator,
    step=1,
    cv=cv_strategy,
    scoring='roc_auc',
    min_features_to_select=10,
    n_jobs=-1,
    verbose=1
)

selector.fit(X, y)

optimal_features = X.columns[selector.support_].tolist()
```

This process selects only 12 features that could optimize our model

```python
len(optimal_features)
```

Here we can see that although we are using way less features than our previous attempt, the model still retain a high level lof ROC AUC. Using less features also helps us in monitoring and maintaining our model performance through PSI (Population Stability Index)

```python
class_1_index = np.where(selector.classes_ == 1)[0][0]

proba_predictions = selector.predict_proba(X)[:, class_1_index]
predictions = selector.predict(X)

final_auc = roc_auc_score(y, proba_predictions)
print(f"Final AUC: {final_auc}")

print('Classification Report')
print(classification_report(y, predictions))
```

# Final Model Training

Finally, after selecting the best performing model, parameters for the model, and features we will be training our model with all of the training data. This ensures that the model have the best performance when used for the testing data

```python
final_lgb_model = lgb.LGBMClassifier(**best_lgb_params)

final_lgb_model.fit(X[optimal_features], y)

class_1_index = np.where(final_lgb_model.classes_ == 1)[0][0]

proba_predictions = final_lgb_model.predict_proba(X[optimal_features])[:, class_1_index]
predictions = final_lgb_model.predict(X[optimal_features])

final_auc = roc_auc_score(y, proba_predictions)
print(f"Final AUC: {final_auc}")

print('Classification Report')
print(classification_report(y, predictions))
```

```python
import cloudpickle

model_path = Path('model')
model_path.mkdir(exist_ok=True)

with open((model_path / 'lgb_model').with_suffix('.pkl'), 'wb') as f:
    cloudpickle.dump(final_lgb_model, f)
```
